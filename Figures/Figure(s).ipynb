```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set font style globally before any plotting
plt.rcParams["font.family"] = "Arial"  # Ensure Times New Roman is applied

# Load the dataset
file_path = 'Dataset 2017-2020.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Check if the column exists in the dataset
column_name = 'Urban_Type'  # Replace with the correct column name
if column_name in data.columns:
    # Calculate percentage distribution
    category_counts = data[column_name].value_counts(normalize=True) * 100  # Convert to percentages
    
    # Generate bar chart
    fig, ax = plt.subplots(figsize=(6, 4.5))  # Adjust figure size as needed
    
    # Bar positions and data
    x_positions = np.arange(len(category_counts))
    bars = ax.bar(x_positions, category_counts.values, color='#DD8452', width=0.4)  # Vibrant orange color

    # Add percentages above bars
    for bar, value in zip(bars, category_counts.values):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{value:.0f}%', 
                ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    # Formatting
    
    ax.set_title('Distribution of Urban - Rural Sections', fontsize=14, fontweight='bold')
    ax.set_xticks(x_positions)
    ax.set_xticklabels(category_counts.index, fontsize=12)
    ax.set_xticklabels([label.capitalize() for label in category_counts.index], fontsize=12)
    ax.set_ylim(0, max(category_counts.values) + 10)  # Adjust y-axis limit for better spacing
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    #plt.tight_layout()

    # Save the chart
    output_file = f'{column_name}_formatted_bar_chart.png'
    #plt.savefig('Urbanity.png', dpi=4000)
    plt.show()
    
else:
    print(f"Column '{column_name}' not found in the dataset.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define figure with two subplots for 2-year and 3-year predictions
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

plt.rcParams["font.family"] = "Arial"

# Bar width
bar_width = 0.35  

# Define model names explicitly
model_names = ["CatBoost", "RF", "XGBoost"]  # Define models directly

# X-axis positions
x = np.arange(len(model_names))  # Positions for each model

# Define colors for Augmented and Non-Augmented versions (same color across models)
no_aug_color = "#1f77b4"  # Blue for No Augmentation
aug_color = "#ff7f0e"  # Orange for With Augmentation

# ✅ Define RMSE values (Make sure these values match your dataset)
rmse_no_aug = {
    "CatBoost": [18.78, 17.88],  # [3-year, 2-year]
    "RF": [21.9, 21.21],
    "XGBoost": [18.8, 17.91],
}

rmse_aug = {
    "CatBoost": [12.13, 12.41],  # [3-year, 2-year]
    "RF": [17.13, 17.93],
    "XGBoost": [12.06, 12.43],
}

# Create bar charts for each subplot
titles = ["3-Year Prediction", "2-Year Prediction"]
for i, ax in enumerate(axes):
    # Get RMSE values for this prediction period (i=0 -> 2-Year, i=1 -> 3-Year)
    rmse_no_aug_values = [rmse_no_aug[model][i] for model in model_names]
    rmse_aug_values = [rmse_aug[model][i] for model in model_names]

    # Plot bars
    bars1 = ax.bar(x - bar_width / 2, rmse_no_aug_values, bar_width, 
                   label="No Augmentation", color=no_aug_color, alpha=0.8)
    bars2 = ax.bar(x + bar_width / 2, rmse_aug_values, bar_width, 
                   label="With Augmentation", color=aug_color, alpha=0.8)

    # Formatting
    ax.set_title(titles[i], fontsize=14, fontweight="bold")
    if i == 0:
        ax.set_ylabel("RMSE", fontsize=12)
    ax.set_xticks(x)
    ax.set_xticklabels(model_names, fontsize=13)
    ax.set_ylim(0, 25)
    ax.grid(axis="y", linestyle="--", alpha=0.6)

    # Add values on top of bars
    for bar in bars1 + bars2:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.2f}', 
                ha='center', va='bottom', fontsize=10)

# Custom Legend (only on the first subplot for clarity)
axes[0].legend(fontsize=12, shadow=False, frameon=True, facecolor='white', edgecolor='black')
axes[1].legend(fontsize=12, shadow=False, frameon=True, facecolor='white', edgecolor='black')

# Adjust layout for clarity
plt.tight_layout()
plt.savefig("Results.png", dpi=1000)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define figure with two subplots for 2-year and 3-year predictions
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

plt.rcParams["font.family"] = "Arial"

# Bar width
bar_width = 0.25  

# Define model names explicitly
model_names = ["CatBoost", "RF", "XGBoost"]

# X-axis positions
x = np.arange(len(model_names))  # Positions for each model

# Define colors for each augmentation method
original_color = "#1f77b4"  # Blue for Original
gaussian_color = "#ff7f0e"  # Orange for Gaussian Noise-Based Augmentation
smote_color = "#2ca02c"  # Green for SMOTE

# ✅ Define WMAP values for each model and augmentation method
wmap_original = {
    "CatBoost": [10.11, 9.13],  # [3-year, 2-year]
    "RF": [12.52, 11.57],
    "XGBoost": [10.26, 9.25],
}

wmap_gaussian = {
    "CatBoost": [6.00, 6.19],  # [3-year, 2-year]
    "RF": [8.89, 9.19],
    "XGBoost": [5.62, 5.75],
}

wmap_smote = {
    "CatBoost": [10.65, 10.29],  # [3-year, 2-year]
    "RF": [11.89, 11.79],
    "XGBoost": [11.45, 10.29],
}

# Create bar charts for each subplot
titles = ["3-Year Prediction", "2-Year Prediction"]
for i, ax in enumerate(axes):
    # Get WMAP values for this prediction period (i=0 -> 3-Year, i=1 -> 2-Year)
    wmap_original_values = [wmap_original[model][i] for model in model_names]
    wmap_gaussian_values = [wmap_gaussian[model][i] for model in model_names]
    wmap_smote_values = [wmap_smote[model][i] for model in model_names]

    # Plot bars
    bars1 = ax.bar(x - bar_width, wmap_original_values, bar_width, 
                   label="Original", color=original_color, alpha=0.8)
    bars2 = ax.bar(x, wmap_gaussian_values, bar_width, 
                   label="Gaussian Noise-Based", color=gaussian_color, alpha=0.8)
    bars3 = ax.bar(x + bar_width, wmap_smote_values, bar_width, 
                   label="SMOTE", color=smote_color, alpha=0.8)

    # Formatting
    ax.set_title(titles[i], fontsize=14, fontweight="bold")
    if i == 0:
        ax.set_ylabel("WMAP (%)", fontsize=12)
    ax.set_xticks(x)
    ax.set_xticklabels(model_names, fontsize=12)
    ax.set_ylim(0, 15)  # Set y-axis range to maintain consistency
    ax.grid(axis="y", linestyle="--", alpha=0.6)

    # Add values on top of bars
    for bars in [bars1, bars2, bars3]:
        for bar in bars:
            ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
                    f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=10)

# Custom Legend (shared across both subplots)
axes[0].legend(fontsize=12, shadow=False, frameon=True, facecolor='white', edgecolor='black')
axes[1].legend(fontsize=12, shadow=False, frameon=True, facecolor='white', edgecolor='black')

# Adjust layout for clarity
plt.tight_layout()
plt.savefig("WMAP_Comparison.png", dpi=1000)
plt.show()
```
